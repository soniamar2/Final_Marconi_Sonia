# Matplotlib
Matplotlib is a very tradictional data visualization library in Python published in 2003. It is the foundation of many other libraries, such as seaborn, pygal, and so on. It is very flexible and customizable, but it is also very verbose and not very efficient, which usually requires extensive code for simple visualizations.
Pros: Foundation for many other libraries; highly customizable.
Cons: Steeper learning curve; can require extensive code for simple visualizations; aesthetics might seem outdated.

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.


```python
from matplotlib import pyplot as plt
```

```python
# simple data visualization
x_values = [1,2,3,4]
y_values = [5,4,6,2]
plt.plot(x_values, y_values)
plt.show()
# we can save the plot etc direclty from it
```

```python
# scatter plot
plt.scatter(x_values,y_values)
plt.show()
```

```python
# we can combine plots on the same figure
plt.scatter(x_values,y_values)
other_x_values = [1,2,3,4]
other_y_values = [4,2,3,9]
plt.plot(other_x_values, other_y_values, color= "red")
# customize our plot
plt.title("sample plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

matplotlib is a data visualization library for Python that provides a wide range of tools and functions for creating graphs. It is highly versatile and can be used to produce a variety of graph types and customizations.

Types of Graphs:

### Line graphs
Useful for representing trends and data changes over time

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.title('Line Graph Example')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

### Histograms: 
Perfect for analyzing the distribution of data sets.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30)
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

### Bar charts: 
Horizontal and vertical, ideal for quantitative comparisons.

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C']
values = [10, 15, 7]

plt.bar(categories, values)
plt.title('Vertical Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()

# For a horizontal bar chart, use plt.barh()
plt.barh(categories, values)
plt.title('Horizontal Bar Chart Example')
plt.xlabel('Values')
plt.ylabel('Category')
plt.show()
```

### Pie charts: 
To show proportions and percentages.

```python
import matplotlib.pyplot as plt

# Data
sizes = [215, 130, 245, 210]
labels = ['Apple', 'Banana', 'Cherry', 'Date']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Scatter plots: 
Useful for exploring relationships or correlations between two variables.

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 2, 3, 4]

plt.scatter(x, y)
plt.title('Scatter Plot Example')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

### Box plots: 
To display statistical distributions and identify outliers.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = np.random.normal(0, 1, 1000)

plt.boxplot(data)
plt.title('Box Plot Example')
plt.ylabel('Values')
plt.show()
```

### Error bars: 
To depict uncertainty around measurements or estimates.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = [1, 2, 3, 4, 5]
y = np.power(x, 2)
errors = np.sqrt(x)

plt.errorbar(x, y, yerr=errors, fmt='-o')
plt.title('Error Bars Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Area plots: 
To highlight accumulated quantities over time.

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.fill_between(x, y)
plt.title('Area Plot Example')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

### Contour plots: 
Used to represent three-dimensional data in two dimensions.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.title('Contour Plot Example')
plt.colorbar()
plt.show()
```

### Surface plots: 
For visualizing three-dimensional data.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Surface plot
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title('Surface Plot Example')
plt.colorbar(surf)
plt.show()
```

## Customization and Styling:

Modify colors, line types, markers, and more to tailor the graph to your desired style.
Ability to add text, annotations, and reference lines to graphs for improved clarity.
Manage legends, titles, and axis labels for better interpretation.
Use color palettes and color maps to enhance visual data representation.
Supports LaTeX for mathematical texts and annotations.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 50)
y = np.sin(x)

# Create a plot with a custom style
plt.figure(figsize=(10, 6))
line, = plt.plot(x, y, label='sin(x)', color='magenta', linewidth=2, linestyle='--', marker='o', markersize=8)

# Adding text and annotation
plt.text(5, 0, 'Middle Point', fontsize=12, ha='center')
plt.annotate('Local Max', xy=(1.5, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=12, color='darkgreen')

# Adding a reference line
plt.axhline(0, color='grey', linewidth=0.5, linestyle='dotted')

# Using LaTeX for mathematical text
plt.title(r'Example of $\sin(x)$ Function', fontsize=16)
plt.xlabel(r'$x$ values', fontsize=14)
plt.ylabel(r'$\sin(x)$', fontsize=14)

# Using color palettes and color maps
plt.set_cmap('viridis')

# Manage legends and axis labels
plt.legend(title='Legend')

# Show the plot
plt.show()
```

## Structure and Layout:

Support for multiple panels (subplots) in a single figure, allowing comparison of different graphs.
Grid and alignment features to organize graphs coherently.
Dynamic layout adjustment to prevent overlapping texts and titles.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# Create a figure and a set of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # 2x2 grid of subplots

# Plotting on each subplot
axs[0, 0].plot(x, y1, 'tab:red')
axs[0, 0].set_title('Sine Wave')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('sin(X)')

axs[0, 1].plot(x, y2, 'tab:blue')
axs[0, 1].set_title('Cosine Wave')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('cos(X)')

axs[1, 0].plot(x, y3, 'tab:green')
axs[1, 0].set_title('Sine * Cosine Wave')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('sin(X) * cos(X)')

# Adjust layout to prevent overlaps
plt.tight_layout()

# Hide the 4th subplot (bottom right) as it's not being used
axs[1, 1].axis('off')

# Show the plot
plt.show()
```

### features:
- multiple panels (subplots): plt.subplots(2,2)
- dynamic layout adjustment: plt.tight_layout() --> it automatically adjusts subplots params so that the subplot fits into the figure area. The empty subplot is turned off with axs[1, 1].axis('off') to maintain a clean layout.

## Interactive Features
matplotlib offers several interactive features that can enhance the way users explore and present their data visualizations, although it's primarily a plotting library rather than a fully interactive visualization tool like some web-based libraries (e.g., Bokeh or Plotly). However, for many applications, especially in scientific and academic environments, matplotlib provides sufficient interactivity. Here's how matplotlib supports interactivity and export:
- Zoom and Pan: When viewing plots in interactive mode (e.g., using IPython or a Jupyter notebook), matplotlib allows users to zoom in and pan across plots. This is particularly useful for detailed inspection of high-density plots.
- Event Handling: matplotlib can handle a variety of events like mouse clicks, key presses, and more. This allows for custom interactivity where actions can be programmed in response to user inputs.
- Dynamic Updates: Plots can be dynamically updated in an interactive session. This is useful for applications that require real-time data updates.

```python
# example di pan
# Ensure to run this in a Jupyter notebook# Using pip

%matplotlib widget 
import matplotlib.pyplot as plt
import numpy as np

# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Interactive Plot with Pan and Zoom')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Show the plot with interactive features
plt.show()
```

## Capability to Save Graphs in Various Formats
matplotlib provides extensive support for exporting graphs to a variety of file formats, including PNG, PDF, SVG, EPS, and many others. This flexibility is crucial for preparing visuals for different mediums, be it web publishing, academic papers, or presentations.
The savefig() function in matplotlib allows for detailed customization of the export process, including setting the DPI (dots per inch), choosing the format, and deciding whether to use tight bounding boxes to optimize space around the plot.

## Integration with GUI Toolkits
Tkinter, PyQt, and other GUI toolkits: matplotlib can be embedded in applications using various GUI toolkits. This allows developers to build custom-tailored graphical applications that include matplotlib plots as integral components.
Navigation Toolbar: When embedded in GUI applications, matplotlib plots can include a navigation toolbar, which provides buttons for zooming and panning, saving the plot to a file, and other interactive features.

A **GUI Toolkit**, or Graphical User Interface Toolkit, is a collection of software libraries that provide the necessary tools for developing applications with graphical interfaces. These toolkits offer a wide range of visual elements, such as windows, buttons, menus, navigation bars, and other widgets that can be used to build interactive and visually appealing user interfaces.

### Key Functions of GUI Toolkits
1. **Widgets**: Provide basic components (called widgets), like buttons, text boxes, labels, etc., essential for constructing functional user interfaces.
2. **Event Handling**: Manage user interactions, such as mouse clicks, key presses, mouse movements, etc., allowing developers to define specific responses to these actions.
3. **Layout and Styling**: Offer options for organizing and styling UI components, helping developers to create cohesive and adaptive designs that work across different devices and screen sizes.
4. **Accessibility**: Many advanced toolkits advance features that make applications accessible, such as support for screen readers, keyboard navigation, and high visual contrast.

### Examples of GUI Toolkits
- **Tkinter**: A commonly used toolkit in Python, suitable for quick scripting and prototyping. It comes bundled with Python and offers a simple interface for creating basic applications.
- **Qt**: Available primarily through its Python bindings PyQt and PySide, it is a powerful and versatile toolkit for developing complex and professional applications that can run on various platforms (Windows, macOS, Linux).
- **GTK**: Used in many open-source projects, including the GNOME desktop environment. It's suitable for Linux applications but also supports other operating systems.
- **WxWidgets**: Another cross-platform toolkit that allows developing applications that look native on the platforms they run on.
- **.NET Framework**: Provides Windows Forms and WPF (Windows Presentation Foundation), two toolkits for developing desktop applications on Windows.
- **Java Swing and JavaFX**: Toolkits for Java that enable graphical interface creation. JavaFX is the more modern of the two and offers more advanced features and better styling.

### Choosing a GUI Toolkit
Selecting a GUI toolkit depends on several factors, including:
- **Target Platform**: Some toolkits are better for specific platforms (e.g., Windows Forms for Windows).
- **Application Complexity**: Toolkits like Qt support very complex and feature-rich applications.
- **Language Preferences**: Some toolkits are tied to specific programming languages.
- **License and Cost**: Some toolkits are open source, others require paid licenses, especially for commercial use.

GUI toolkits are essential for modern software development, facilitating the creation of applications that users can easily navigate and use.
